# Plot UMNN 2D dataset results

```python
reps = 100
nb_samps = np.logspace(3,4,5)
nb_samps = ((nb_samps//10)*10).astype(int)
ns = np.size(nb_samps)

toys = ["8gaussians1","8gaussians2", 
        "swissroll1","swissroll2", 
        "moons1","moons2",
        "pinwheel1","pinwheel2",
        "2spirals1","2spirals2", 
        "checkerboard1","checkerboard2",
        "circles1","circles2",
        "joint_gaussian1","joint_gaussian2"]

results = np.zeros([len(toys),reps,ns])
```

```python
toy_ct = 0
for toy in toys:
    for rep in range(reps):
        file = open("results/umnnaf_results/{}_{}.out".format(toy,rep+1), "r").read()
        file=file.replace('[','')
        file=file.replace(']\n','')
        results[toy_ct,rep,:] = [float(x) for x in file.split()]
    
    toy_ct += 1
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['mathtext.rm'] = 'Bitstream Vera Sans'
plt.rcParams['mathtext.it'] = 'Bitstream Vera Sans:italic'
plt.rcParams['mathtext.bf'] = 'Bitstream Vera Sans:bold'

toys = ["8gaussians","swissroll","moons","pinwheel","2spirals","checkerboard",
        "circles","joint_gaussian"]

toy_ct = 0
for toy in toys:
    plt.figure()
    
    plt.loglog(nb_samps,np.mean(results[2*toy_ct,:,:],axis=0),alpha=0,color="black")
    plt.loglog(nb_samps,np.mean(results[2*toy_ct+1,:,:],axis=0),alpha=0,color="black")

    plt.errorbar(nb_samps,np.mean(results[2*toy_ct,:,:],axis=0),linewidth=4,
                yerr = 1.96*np.std((results[2*toy_ct,:,:])/np.sqrt(reps),axis=0),
                label = "horizontal".format(toy), alpha=1)
    plt.errorbar(nb_samps,np.mean(results[2*toy_ct+1,:,:],axis=0),linewidth=4,
                yerr = 1.96*np.std((results[2*toy_ct+1,:,:])/np.sqrt(reps),axis=0),
                label = "vertical".format(toy), alpha=1)

    plt.ylabel("negative log-likelihood", fontsize=18)
    plt.xlabel(r"sample size", fontsize=18)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.legend(fontsize=16)

    plt.tight_layout()
    plt.title("{}".format(toy))
    
    plt.savefig('figures/{}_order.pdf'.format(toy))
    
    toy_ct += 1
```

# Plot 2D sine density results

```python
reps = 100
nb_samps = np.logspace(3,4,5)
nb_samps = ((nb_samps//10)*10).astype(int)
ns = np.size(nb_samps)

toys = ['sine13','sine31','sine15','sine51','sine17','sine71']

results = np.zeros([len(toys),reps,ns])
```

```python
toy_ct = 0
for toy in toys:
    for rep in range(reps):
        file = open("results/sine_product_results/{}_{}.out".format(toy,rep+1), "r").read()
        file=file.replace('[','')
        file=file.replace(']\n','')
        results[toy_ct,rep,:] = [float(x) for x in file.split()]
    
    toy_ct += 1
```

```python
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman'] + plt.rcParams['font.serif']
plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['mathtext.rm'] = 'Bitstream Vera Sans'
plt.rcParams['mathtext.it'] = 'Bitstream Vera Sans:italic'
plt.rcParams['mathtext.bf'] = 'Bitstream Vera Sans:bold'

toys = ["sine13", "sine15", "sine17"]

toy_ct = 0

cmap = plt.get_cmap("tab10")

plt.figure()

for toy in toys:
    plt.loglog(nb_samps,np.mean(results[2*toy_ct,:,:],axis=0),alpha=0,color=cmap(toy_ct))
    plt.loglog(nb_samps,np.mean(results[2*toy_ct+1,:,:],axis=0),alpha=0,color=cmap(toy_ct))
 
    plt.errorbar(nb_samps,np.mean(results[2*toy_ct,:,:],axis=0),linewidth=4,
                yerr = 1.96*np.std((results[2*toy_ct,:,:])/np.sqrt(reps),axis=0),
                alpha=1, linestyle = ':', color = cmap(toy_ct))
    plt.errorbar(nb_samps,np.mean(results[2*toy_ct+1,:,:],axis=0),linewidth=4,
                yerr = 1.96*np.std((results[2*toy_ct+1,:,:])/np.sqrt(reps),axis=0),color=cmap(toy_ct),
                label = (r"$k_2={}$").format(2*toy_ct+3),
                alpha=1, linestyle = '-')
    
    toy_ct += 1
    
plt.ylabel("negative log-likelihood", fontsize=18)
plt.xlabel(r"sample size", fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(fontsize=16)

plt.tight_layout()
plt.title("{}".format(toy))
    
plt.savefig('figures/sine_product_order.pdf')
```

# Density heatmaps

```python
datasets = ["8gaussians1",
            "swissroll1",
            "moons1",
            "2spirals1",
            "checkerboard1",
            "circles1",
            "joint_gaussian1",
            "mixture_1","mixture_2","mixture_3",
            "sine13","sine31","sine15","sine51","sine17","sine71"]

nb_samp = 100000
device="cpu"

for toy in datasets:
    plt.figure()
    x = toy_data.inf_train_gen(toy, batch_size=nb_samp)
    if toy in ["joint_gaussian1","joint_gaussian2","mixture_1","mixture_2","mixture_3",
              "sine13","sine31","sine15","sine51","sine17","sine71"]:
        x = x.numpy()
    plt.hist2d(x[:,0],x[:,1], bins=200, rasterized=True);
    if toy == "joint_gaussian1":
        plt.xlim([-1.5,3.5])
        plt.ylim([-3,3])
    plt.axis('off')
    
    plt.tight_layout()
    plt.title("{}".format(toy))
    
    plt.savefig('figures/{}_heatmap.pdf'.format(toy))
```
